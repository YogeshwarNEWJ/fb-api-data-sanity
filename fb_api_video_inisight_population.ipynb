### Task 3: Check if the videos published before 45 days have 45 observations or not

```python
from datetime import datetime, timedelta
```

```python
# check the date, before which any video published should have 45 days record
data['published_time_ist'].max() - timedelta(days = 45)
```

```python
# create a dataframe of the videos for which data should be populated 
df_must_45_days = data[data['published_time_ist'] < data['published_time_ist'].max() - timedelta(days = 45)]
```

```python
df_must_45_days['published_time_ist'].max() # max date
```

```python
Counter(df_must_45_days['videoId'])
```

```python
index_id_less_than_45 = [i for i in range(len(list(Counter(df_must_45_days['videoId']).values()))) if list(Counter(df_must_45_days['videoId']).values())[i]<45]
```

```python
ids_with_less_than_45 = [list(Counter(df_must_45_days['videoId']).keys())[i] for i in index_id_less_than_45]
```

```python
len(ids_with_less_than_45)
```

`There are total 185 videos for which data isn't retreived for 45 days`
